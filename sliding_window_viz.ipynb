```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters
sampling_rate = 128  # Hz
duration = 1  # seconds
window_size = 32  # samples
overlap = 8  # samples
step_size = window_size - overlap

# Set seed for reproducibility
np.random.seed(42)

# Generate a sample EEG signal (1280 samples for duration)
eeg_signal = np.random.randn(sampling_rate * duration)

# Function to apply sliding window
def sliding_window(signal, window_size, step_size):
    return [signal[i:i+window_size] for i in range(0, len(signal) - window_size + 1, step_size)]

# Apply sliding window
windows = sliding_window(eeg_signal, window_size, step_size)

# Visualization
plt.figure(figsize=(15, 5))
plt.plot(eeg_signal, linewidth=3)
for i, window in enumerate(windows):
    start = i * step_size
    end = start + window_size
    plt.axvspan(start, end, color='orange', alpha=0.2)

# Increase sizes of title, labels, legend, ticks and line
plt.rc('font', weight='normal')
plt.rc('axes', titlesize=30)
plt.rc('axes', labelsize=30)
plt.rc('xtick', labelsize=40)
plt.rc('ytick', labelsize=40)
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Times New Roman"]

# Remove y-axis ticks and label
plt.yticks([])
plt.ylabel('')
#plt.ylabel('EEG signal amplitude')

# Set x-axis range and ticks from 0 to 128 (inclusive)
plt.xlabel('Sample number')
plt.xlim(0, 128)
plt.xticks(np.arange(0, 128 + 1, 32))
plt.tick_params(axis='x', width=2, length=10)

plt.show()
```
